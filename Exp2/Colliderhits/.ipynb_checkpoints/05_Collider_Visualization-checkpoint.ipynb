```python
Categories_Sum = pd.DataFrame(complete_data.groupby(["Collider_Categorical"])["SubjectID"].count())
Categories_Sum.reset_index(inplace=True)
Categories_Sum.rename(columns={"SubjectID": "Count"}, inplace=True)
```

```python
Categories_Sum.sort_values(by="Count", ascending=False)
```

```python
Collider_Count = pd.DataFrame(complete_data.groupby(["SubjectID","Session", "Collider_Categorical" ])["SessionSubsection"].count())
Collider_Count.reset_index(inplace=True)
Collider_Count.rename(columns={"SessionSubsection": "Count"}, inplace=True)
```

```python
Collider_Count.head(10)
```

```python
sns.catplot(data=Collider_Count,  y="Count", x="Collider_Categorical", hue="Session", height=6, aspect=3, kind="swarm", dodge=True)
```

```python
sns.catplot(data=Collider_Count[(Collider_Count.Collider_Categorical != "Background") & (Collider_Count.Collider_Categorical != "Building") ],  y="Count", x="Collider_Categorical", hue="Session", height=6, aspect=3, kind="swarm", dodge=True)
```

```python
sns.catplot(data=Collider_Count[(Collider_Count.Collider_Categorical == "Passive_Agent") | (Collider_Count.Collider_Categorical == "Active_Agent")  ],  y="Count", x="Session", hue="Collider_Categorical", height=6, aspect=3, kind='box')
sns.swarmplot(data=Collider_Count[(Collider_Count.Collider_Categorical == "Passive_Agent") | (Collider_Count.Collider_Categorical == "Active_Agent")  ],  y="Count", x="Session",hue="Collider_Categorical", size=6, dodge=True, palette="dark")
```

```python
sns.catplot(data=Collider_Count[(Collider_Count.Collider_Categorical == "TaskBuilding_Public") | (Collider_Count.Collider_Categorical == "TaskBuilding_Residential") ],  y="Count", x="Session", hue="Collider_Categorical", height=6, aspect=3, kind='box')
sns.swarmplot(data=Collider_Count[(Collider_Count.Collider_Categorical == "TaskBuilding_Public") | (Collider_Count.Collider_Categorical == "TaskBuilding_Residential") ],  y="Count", x="Session", hue="Collider_Categorical", size=6, dodge=True, palette="dark")
```

```python
sns.displot(complete_data[(complete_data.Eucledian_distance < 200) & (complete_data.Collider_Categorical != "Background") ], x="Eucledian_distance", hue="Collider_Categorical",  multiple="stack", kde=True, height=6, aspect=3)
```

```python
sns.displot(complete_data, x="Eucledian_distance", hue="Collider_Categorical", height=6, aspect=3)
sns.displot(complete_data[complete_data.Eucledian_distance < 200], x="Eucledian_distance", hue="Collider_Categorical", height=6, aspect=3)
```

```python
sns.relplot(x="Eucledian_distance", y="Continuous_Time", hue="Collider_Categorical", 
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=complete_data[complete_data.Collider_Categorical != "Background"])
```

```python
sns.jointplot(data=complete_data[complete_data.Collider_Categorical != "Background"], x="Eucledian_distance", y="Continuous_Time", kind="kde")
```
