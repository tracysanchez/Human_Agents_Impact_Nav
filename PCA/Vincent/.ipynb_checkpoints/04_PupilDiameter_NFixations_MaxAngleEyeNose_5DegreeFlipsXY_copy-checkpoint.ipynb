```python
plt.hist(df_validData.pupilDiameterMillimetersRight, bins=100)
print('Mean:',np.mean(df_validData.pupilDiameterMillimetersRight))
print('SD:',np.std(df_validData.pupilDiameterMillimetersRight))
print('Max:',max(df_validData.pupilDiameterMillimetersRight))
print('Min:',min(df_validData.pupilDiameterMillimetersRight))
plt.title('Distribution of valid pupil diameter samples (right eye, mm)')
plt.show()
```

```python
# Meadian, SID & quantiles
SIDs = []
meds = []
p25s = []
p75s = []

meds_l = []
p25s_l = []
p75s_l = []

for subject in np.unique(df_validData.SID):
    df_subj = df_validData[df_validData['SID'] == subject]
    p_r_med = np.nanmedian(df_subj.pupilDiameterMillimetersRight)
    p_r_25p = np.quantile(df_subj.pupilDiameterMillimetersRight, 0.25)
    p_r_75p = np.quantile(df_subj.pupilDiameterMillimetersRight, 0.75)
    SIDs.append(subject)
    meds.append(p_r_med)
    p25s.append(p_r_25p)
    p75s.append(p_r_75p)

    p_l_med = np.nanmedian(df_subj.pupilDiameterMillimetersLeft)
    p_l_25p = np.quantile(df_subj.pupilDiameterMillimetersLeft, 0.25)
    p_l_75p = np.quantile(df_subj.pupilDiameterMillimetersLeft, 0.75)
    meds_l.append(p_l_med)
    p25s_l.append(p_l_25p)
    p75s_l.append(p_l_75p)
    
df_pupilD = pd.DataFrame()
df_pupilD['SID'] = SIDs
df_pupilD['pd_r_mm_25%'] = p25s
df_pupilD['pd_l_mm_25%'] = p25s_l
df_pupilD['pd_r_mm_med'] = meds
df_pupilD['pd_l_mm_med'] = meds_l
df_pupilD['pd_r_mm_75%'] = p75s
df_pupilD['pd_l_mm_75%'] = p75s_l
df_pupilD
```

```python
# Calculate distance between 25th perceptile and 75th percentile & the Bowley coefficient to account for skew
df_pupilD['pd_r_mm_DistQ1-Q3'] = df_pupilD['pd_r_mm_75%'] - df_pupilD['pd_r_mm_25%']
df_pupilD['pd_l_mm_DistQ1-Q3'] = df_pupilD['pd_l_mm_75%'] - df_pupilD['pd_l_mm_25%']
# BOWLEY COEFFICIENT: ranges from -1 to 1, with 0 indicating a symmetrical distribution, negative values indicating a left skew, and positive values indicating a right skew
df_pupilD['pd_r_mm_bowleyCoef'] = (df_pupilD['pd_r_mm_75%'] + df_pupilD['pd_r_mm_25%'] - 2*df_pupilD['pd_r_mm_med']) / df_pupilD['pd_r_mm_DistQ1-Q3']
df_pupilD['pd_l_mm_bowleyCoef'] = (df_pupilD['pd_l_mm_75%'] + df_pupilD['pd_l_mm_25%'] - 2*df_pupilD['pd_l_mm_med']) / df_pupilD['pd_l_mm_DistQ1-Q3']
df_pupilD
```

```python
# drop columns not used in further analysis:
df_pupilD = df_pupilD.drop(['pd_r_mm_25%', 'pd_l_mm_25%', 'pd_r_mm_75%', 'pd_l_mm_75%'], axis=1)
df_pupilD
```

# Max Angle between eye and nose vector (within each fixation) 

```python
# load in df
start = time.time()
df = pd.read_csv(r"F:\WestbrookProject\SpaRe\control_group\pre-processing_2023\velocity_based\GazeProcessedMerge_correctedFixations_15_02_25.csv")
end = time.time()
print(end-start)
df
```

```python
# Filter fixations
df_fix = df[df['isFix'].notna()]
df_fix
```

```python
# Checking that there are no errors (i.e., two endings or starts of a FIXATION  following each other.
df_fix2 = df_fix[df_fix['events'].notna()]
df_fix2['events']

values = []
nex_values=[]
bools = []
idxs = []
gidxs = []
participants = []
sessions=[]
parts=[]
for participant in np.unique(df_fix2.SID):
    df_sid = df_fix2[df_fix2['SID'] == participant]
    for session in np.unique(df_sid.Session):
        df_ses = df_sid[df_sid['Session'] == session]
        #print("Starting session:", session)
        for part in np.unique(df_ses.ET):
            df_part = df_ses[df_ses['ET'] == part]
            df_part = df_part.reset_index(drop=True)
            for i in range(len(df_part)-1):
                cur = df_part['events'][i] 
                nex = df_part['events'][i+1]
                bool1 = cur == nex
                bools.append(bool1)
                idxs.append(i)
                gidxs.append(df_part['Unnamed: 0'][i])
                values.append(cur)
                nex_values.append(nex)
                participants.append(participant)
                sessions.append(session)
                parts.append(part)
DoubleEndDF = pd.DataFrame()
DoubleEndDF['bools'] = bools
DoubleEndDF['idxs'] = idxs
DoubleEndDF['gidxs'] = gidxs
DoubleEndDF['value'] = values
DoubleEndDF['sid'] = participants
DoubleEndDF['ses'] = sessions
DoubleEndDF['part'] = parts
print(DoubleEndDF[DoubleEndDF['bools']==True])
print(DoubleEndDF)
#print('Samples starting:',len(DoubleEndDF[DoubleEndDF['value']==2]))
#print('Samples ending:',len(DoubleEndDF[DoubleEndDF['value']==-2]))
```

```python
# Calculating combined eye and nose vectors:
res=[]
LocalX = df_fix.eyeDirectionCombinedLocal_x
LocalY = df_fix.eyeDirectionCombinedLocal_y
LocalZ = df_fix.eyeDirectionCombinedLocal_z
NoseVector = np.array([0, 0, 1])
LocalGazeList = list(map(list, zip(LocalX, LocalY, LocalZ)))
LocalGazeList
LocalGazeArrays=np.array(LocalGazeList)
LocalGazeArrays
```

```python
def unit_vector(vector):
    return vector / np.linalg.norm(vector)
def angle_between(v1, v2):
    # Returns the angle in radians between vectors 
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))

results=[]
for i in range(len(LocalGazeArrays)):
    res = math.degrees(angle_between(LocalGazeArrays[i], NoseVector))
    results.append(res) 
df_fix['GazeAngle']=results
df_fix
```

```python
# taking max angle to nose vector for each fixation:
values = []

max_values = []
idxs=[]
gidxs = []
participants = []
sessions=[]
parts=[]
for participant in np.unique(df_fix.SID):
    print(participant)
    df_sid = df_fix[df_fix['SID'] == participant]
    for session in np.unique(df_sid.Session):
        df_ses = df_sid[df_sid['Session'] == session]
        #print("Starting session:", session)
        for part in np.unique(df_ses.ET):
            df_part = df_ses[df_ses['ET'] == part]
            df_part = df_part.reset_index(drop=True)
            for i in range(len(df_part)-1):
                cur = df_part['events'][i]
                values.append(df_part['GazeAngle'][i])
                if cur == -2:
                    # only at the end of a fixation, calculate maximum
                    maxv = np.nanmax(values)
                    max_values.append(maxv)
                    gidxs.append(df_part['index'][i])
                    idxs.append(i)
                    participants.append(participant)
                    sessions.append(session)
                    parts.append(part)
                    # reset values
                    values = []

df_MaxFix = pd.DataFrame()
df_MaxFix['max_angle'] = max_values
df_MaxFix['idxs'] = idxs
df_MaxFix['gidxs'] = gidxs
df_MaxFix['sid'] = participants
df_MaxFix['ses'] = sessions
df_MaxFix['part'] = parts
df_MaxFix
```

```python
plt.hist(df_MaxFix.max_angle, bins=100)
print('Mean:',np.mean(df_MaxFix.max_angle))
print('SD:',np.std(df_MaxFix.max_angle))
print('Max:',max(df_MaxFix.max_angle))
print('Min:',min(df_MaxFix.max_angle))
plt.title('Distribution of maximum angle between eye and nose vector during fixations in °')
plt.show()
```

```python
# Meadian, SID & quantiles
SIDs = []
meds = []
p25s = []
p75s = []

for subject in np.unique(df_MaxFix.sid):
    df_subj = df_MaxFix[df_MaxFix['sid'] == subject]
    med = np.nanmedian(df_subj.max_angle)
    p25 = np.quantile(df_subj.max_angle, 0.25)
    p75 = np.quantile(df_subj.max_angle, 0.75)
    SIDs.append(subject)
    meds.append(med)
    p25s.append(p25)
    p75s.append(p75)
    
df_Ang = pd.DataFrame()
df_Ang['SID'] = SIDs
df_Ang['MaxAng_EyeNose_25%'] = p25s
df_Ang['MaxAng_EyeNose_med'] = meds
df_Ang['MaxAng_EyeNose_75%'] = p75s
df_Ang
```

```python
# Calculate distance between 25th perceptile and 75th percentile & the Bowley coefficient to account for skew
df_Ang['MaxAng_EyeNose_DistQ1-Q3'] = df_Ang['MaxAng_EyeNose_75%'] - df_Ang['MaxAng_EyeNose_25%']
# BOWLEY COEFFICIENT: ranges from -1 to 1, with 0 indicating a symmetrical distribution, negative values indicating a left skew, and positive values indicating a right skew
df_Ang['pd_r_mm_bowleyCoef'] = (df_Ang['MaxAng_EyeNose_75%'] + df_Ang['MaxAng_EyeNose_25%'] - 2*df_Ang['MaxAng_EyeNose_med']) / df_Ang['MaxAng_EyeNose_DistQ1-Q3']

# drop columns not used in further analysis:
df_Ang = df_Ang.drop(['MaxAng_EyeNose_25%', 'MaxAng_EyeNose_75%'], axis=1)
df_Ang
```

```python
###SAVING###
df_Ang.to_csv(r"C:\Users\Vincent\Documents\Dynamic_VS_Static_Viewing\FinalAnalysis\Preprocessing_ET\ResultingDataFrames\MaxAngleEyeNoseVectorDuringFixation.csv")
```

# FLIPS / Count of eye rotations larger than 5° and count of crossing middle point of X/Y-axis of screen.
### Here we count how often subjects' gaze was changing direction by 5° or more

```python
# reload dataframe if needed
start = time.time()
df = pd.read_csv(r"F:\WestbrookProject\SpaRe\control_group\pre-processing_2023\velocity_based\GazeProcessedMerge_correctedFixations_15_02_25.csv")
end = time.time()
print(end-start)
df
```

```python
# Filter out NaN values
df_fix = df[df['eyeDirectionCombinedLocal_x'].notna()]
NoseVector = np.array([0, 0, 1])
# dropping duplicate columns from resetting the index
df_fix=df_fix.drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.2'],axis=1)
# removing invalid samples
df_fix = df_fix[df_fix['eyeDirectionCombinedLocal_z'] >0.7] # removing impossible angles
df_fix
```

```python
# Calculating combined eye vectors:
LocalX = df_fix.eyeDirectionCombinedLocal_x
LocalY = df_fix.eyeDirectionCombinedLocal_y
LocalZ = df_fix.eyeDirectionCombinedLocal_z
SidewardsX = list(map(list, zip(LocalX, LocalZ)))
UpwardsY = list(map(list, zip(LocalY, LocalZ)))
```

```python
# Creating function that calculates the angle between two vectors v1 and v2
def angle_between(p1, p2):
    ang1 = np.arctan2(*p1[::-1])
    ang2 = np.arctan2(*p2[::-1])
    return np.rad2deg((ang1 - ang2) % (2 * np.pi))
```

```python
# Sidewards / X
results=[]
for i in range(1,len(SidewardsX)):
    res = angle_between(SidewardsX[i-1], SidewardsX[i])
    results.append(res) 
results.append(0)
```

```python
df_fix = df_fix.reset_index(drop=True)
results_2 = []
for i in range(len(df_fix)):
    if (results[i] > 180):
        results_2.append(results[i]-360)
    else:
        results_2.append(results[i])
df_fix['SidewardsAngle_X']=results_2
df_fix
```

```python
# upwards / Y
results=[]
for i in range(1,len(UpwardsY)):
    res = angle_between(UpwardsY[i-1], UpwardsY[i])
    results.append(res) 
results.append(0)
```

```python
df_fix = df_fix.reset_index(drop=True)
results_2 = []
for i in range(len(df_fix)):
    if (results[i] > 180):
        results_2.append(results[i]-360)
    else:
        results_2.append(results[i])
df_fix['UpwardsAngle_Y']=results_2
df_fix
```

## Y-Axis (Up / down on screen)

```python
sum_angles = 0

angles_list = []
idxs=[]
gidxs = []
participants = []
sessions=[]
parts=[]

for participant in np.unique(df_fix.SID):
    print(participant)
    df_sid = df_fix[df_fix['SID'] == participant]
    for session in np.unique(df_sid.Session):
        df_ses = df_sid[df_sid['Session'] == session]
        #print("Starting session:", session)
        for part in np.unique(df_ses.ET):
            df_part = df_ses[df_ses['ET'] == part]
            df_part = df_part.reset_index(drop=True)
            for i in range(len(df_part)-1):
                sum_angles = sum_angles + df_part['UpwardsAngle_Y'][i]
                if sum_angles >= 5 or sum_angles <= -5:
                    angles_list.append(sum_angles)
                    sum_angles = 0 # reset angle to 0
                    gidxs.append(df_part['index'][i])
                    idxs.append(i)
                    participants.append(participant)
                    sessions.append(session)
                    parts.append(part)

df_5gedFlip = pd.DataFrame()
df_5gedFlip['UpwardsAngle_Y'] = angles_list
df_5gedFlip['idxs'] = idxs
df_5gedFlip['gidxs'] = gidxs
df_5gedFlip['sid'] = participants
df_5gedFlip['ses'] = sessions
df_5gedFlip['part'] = parts
df_5gedFlip
```

```python
# Here we remove all rotations larger than 5° that were going into the same direction as the previous rotation
angles_list = []
idxs=[]
gidxs = []
participants = []
sessions=[]
parts=[]

for i in range(len(df_5gedFlip['UpwardsAngle_Y'])-1):
    cur = df_5gedFlip['UpwardsAngle_Y'][i]
    nex = df_5gedFlip['UpwardsAngle_Y'][i+1]
    if np.sign(cur) != np.sign(nex):
        angles_list.append(cur)
        gidxs.append(df_5gedFlip['gidxs'][i])
        idxs.append(df_5gedFlip['idxs'][i])
        participants.append(df_5gedFlip['sid'][i])
        sessions.append(df_5gedFlip['ses'][i])
        parts.append(df_5gedFlip['part'][i])
df_5gedFlip = pd.DataFrame()
df_5gedFlip['UpwardsAngle_Y'] = angles_list
df_5gedFlip['idxs'] = idxs
df_5gedFlip['gidxs'] = gidxs
df_5gedFlip['sid'] = participants
df_5gedFlip['ses'] = sessions
df_5gedFlip['part'] = parts
df_5gedFlip 
```

```python
# count flips per participant
nr_flips = []
partics = []

for participant in np.unique(df_5gedFlip.sid):
    df_sid = df_5gedFlip[df_5gedFlip['sid'] == participant]
    nr_flip = len(df_sid)-1
    nr_flips.append(nr_flip)
    partics.append(participant)

OverallFlips = pd.DataFrame()
OverallFlips['sid'] = partics
OverallFlips['nr_flips_y'] = nr_flips
OverallFlips
```
