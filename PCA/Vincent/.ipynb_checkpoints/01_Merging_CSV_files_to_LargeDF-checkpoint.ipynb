```python
# Loading in the first half of csv files, creating new columns and flattening the files to common df format
durations = []
recordings = []
lengths = []
half_df_1 = pd.DataFrame()

for filename in glob.glob(path):
    data = pd.read_csv(filename)
    data['SID'] = filename[29:33]
    data['Session'] = filename[42]
    data['ET'] = filename[47]
    print(filename)
    half_df_1 = pd.concat([half_df_1, data])
half_df_1
```

```python
# Sorting (can be skippe if already sorted)
half_df_1 = half_df_1.sort_values(by=['SID', 'Session', 'ET', 'timeStampRS'], ascending=[True, True, True, True])
half_df_1
```

```python
# Loading in the csv files, creating metadata df, and flattening the files to common df format
path = "step3_gazeProcessing_2ndHalf/*.csv" # set folder containing json files you want to process
glob.glob(path)

durations = []
recordings = []
lengths = []
half_df_2 = pd.DataFrame()

for filename in glob.glob(path):
    data = pd.read_csv(filename)
    data['SID'] = filename[29:33]
    data['Session'] = filename[42]
    data['ET'] = filename[47]
    print(filename)
    half_df_2 = pd.concat([half_df_2, data])
half_df_2
```

```python
# Merging both halves to full dataframe:
full_df = pd.concat([half_df_1, half_df_2])
full_df
```

```python
# Use this once to create a large dataframe
chunk_size = 500000
total_rows = len(full_df)
with tqdm(total=total_rows, desc = "Writing") as pbar:
    for i in range(0, total_rows, chunk_size):
        full_df.iloc[i:i+chunk_size].to_csv(r'E:\WestbrookProject\SpaRe\control_group\pre-processing_2023\velocity_based\GazeProcessedMerge.csv', mode ='a', index=False, header=(i==0))
        pbar.update(chunk_size)
print("Processing completed")
```
