# Distance

```python
df.avg_dist
```

```python
fixationSta = df[df['events']==2]
fixationEnd = df[df['events']==-2]
# Plotting:
plt.hist(fixationSta.avg_dist, bins=100)
plt.title('Average distance to fixated object (count per fixation)')
print('Maximum distance to fixated object is:', max(fixationSta.avg_dist), 'unity units / meters')
```

```python
fixationSta_High = fixationSta[fixationSta['avg_dist'] > 350]
plt.hist(fixationSta_High.avg_dist, bins=100)
plt.title('Fixations with a distance above 350 meters')
plt.show()
# There is a very interesting outlier at 1261.6meters
```

```python
# Meadian, SID & quantiles
SIDs = []
meds = []
p25s = []
p75s = []

for subject in np.unique(fixationSta.SID):
    df_subj = fixationSta[fixationSta['SID'] == subject]
    med = np.nanmedian(df_subj.avg_dist)
    p25 = np.nanquantile(df_subj.avg_dist, 0.25)
    p75 = np.nanquantile(df_subj.avg_dist, 0.75)
    SIDs.append(subject)
    meds.append(med)
    p25s.append(p25)
    p75s.append(p75)
    
df_dist = pd.DataFrame()
df_dist['SID'] = SIDs
df_dist['FixationDist_25%'] = p25s
df_dist['FixationDist_med'] = meds
df_dist['FixationDist_75%'] = p75s
df_dist
```

# Angular Velocity

```python
df.validData = df.validData.astype(bool)
df_validData = df[df['validData']==True]
print(len(df_validData))
df_validData
```

```python
# Correcting for microjitter from resampling to constant rate:
df_validData.loc[df_validData['combined_vel'] < 0, 'combined_vel'] = 0
# removing rows with no SID
df_validData = df_validData[df_validData['SID'] > 0]
df_validData
```

```python
# Plotting:
plt.hist(df_validData.combined_vel, bins=100)
plt.title('Distribution of combined angular velocity (all samples)')
print('Maximum angular velocity is:', np.nanmax(df_validData.combined_vel), 'radian / second')
```

```python
high_combined_vel = df_validData[df_validData['combined_vel'] > 300]
plt.hist(high_combined_vel.combined_vel, bins=100)
plt.title('Combined angular velocitites above 350 radian/s')
plt.show()
```

```python
print("Mean combined angular velocity (in r/s):", np.nanmean(df_validData['combined_vel']))
print("SD of combined angular velocity (in r/s):", np.nanstd(df_validData['combined_vel']))
print("Median combined angular velocity (in r/s):", np.nanmedian(df_validData['combined_vel']))
print("Maximum of combined angular velocity (in r/s):", np.nanmax(df_validData['combined_vel']))
print("Minimum of combined angular velocity (in r/s):", np.nanmin(df_validData['combined_vel']))
```

```python
# Median, SID & quantiles
SIDs = []
meds = []
p25s = []
p75s = []

for subject in np.unique(df_validData.SID):
    df_subj = df_validData[df_validData['SID'] == subject]
    med = np.nanmedian(df_subj.combined_vel)
    p25 = np.nanquantile(df_subj.combined_vel, 0.25)
    p75 = np.nanquantile(df_subj.combined_vel, 0.75)
    SIDs.append(subject)
    meds.append(med)
    p25s.append(p25)
    p75s.append(p75)
    
df_AngVel = pd.DataFrame()
df_AngVel['SID'] = SIDs
df_AngVel['Combined_AngVel_25%'] = p25s
df_AngVel['Combined_AngVel_med'] = meds
df_AngVel['Combined_AngVel_75%'] = p75s
df_AngVel
```

```python
# Calculate distance between 25th perceptile and 75th percentile & the Bowley coefficient to account for skew
df_AngVel['Combined_AngVel_DistQ1-Q3'] = df_AngVel['Combined_AngVel_75%'] - df_AngVel['Combined_AngVel_25%']
# BOWLEY COEFFICIENT: ranges from -1 to 1, with 0 indicating a symmetrical distribution, negative values indicating a left skew, and positive values indicating a right skew
df_AngVel['pd_r_mm_bowleyCoef'] = (df_AngVel['Combined_AngVel_75%'] + df_AngVel['Combined_AngVel_25%'] - 2*df_AngVel['Combined_AngVel_med']) / df_AngVel['Combined_AngVel_DistQ1-Q3']

# drop columns not used in further analysis:
df_AngVel = df_AngVel.drop(['Combined_AngVel_25%', 'Combined_AngVel_75%'], axis=1)
df_AngVel
```
