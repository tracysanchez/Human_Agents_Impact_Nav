```python
import pandas as pd
# glob module for wildcards in paths
import glob
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import time
import math
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
start = time.time()
df = pd.read_csv(r"F:\WestbrookProject\SpaRe\control_group\pre-processing_2023\velocity_based\GazeProcessedMerge_correctedFixations_15_02_25.csv")
end = time.time()
print(end-start)
df
```

```python
SaccadeSta = df[df['events']==1]
SaccadeEnd = df[df['events']==-1]
SaccadeEnd
```

```python
SaccadeEnd = SaccadeEnd[SaccadeEnd['length']!=0] # deleting all 0-duration saccades, if any
SaccadeEnd = SaccadeEnd[SaccadeEnd['length']!=np.nan] # deleting all NaN saccades, if any
print("Nr. of SaccadeEnd:", len(SaccadeEnd.length)) 
print('Number of SaccadeSta:', len(SaccadeSta.length))
print("Max:",max(SaccadeEnd.length))
print('Min',min(SaccadeEnd.length))
print("Mean",np.mean(SaccadeEnd.length))
print('Median',np.nanmedian(SaccadeEnd.length))
print('SD:',np.std(SaccadeEnd.length))
print('MAD:',stats.median_abs_deviation(SaccadeEnd.length, scale=1, nan_policy='omit'))
```

```python
plt.hist(SaccadeEnd.length, bins=100)
print(max(SaccadeEnd.length))
plt.show()
```

```python
SaccadeEnd.loc[:,'length']#.astype('int64')
VeryLongSacs = SaccadeEnd[SaccadeEnd.loc[:,'length']>1.5]
plt.hist(VeryLongSacs.length, bins=30)
print(max(VeryLongSacs.length))
plt.title('Count of fixation length above 1.5sec:')
plt.show()
```

```python
# Meadian, SID & quantiles
SIDs = []
meds = []
p25s = []
p75s = []

for subject in np.unique(SaccadeEnd.SID):
    df_subj = SaccadeEnd[SaccadeEnd['SID'] == subject]
    med = np.nanmedian(df_subj.length)
    p25 = np.nanquantile(df_subj.length, 0.25)
    p75 = np.nanquantile(df_subj.length, 0.75)
    SIDs.append(subject)
    meds.append(med)
    p25s.append(p25)
    p75s.append(p75)
    
df_SacDur = pd.DataFrame()
df_SacDur['SID'] = SIDs
df_SacDur['Saccade_Duration_25%'] = p25s
df_SacDur['Saccade_Duration_med'] = meds
df_SacDur['Saccade_Duration_75%'] = p75s
df_SacDur
```

```python
# Calculate distance between 25th perceptile and 75th percentile & the Bowley coefficient to account for skew
df_SacDur['Saccade_Duration_DistQ1-Q3'] = df_SacDur['Saccade_Duration_75%'] - df_SacDur['Saccade_Duration_25%']
# BOWLEY COEFFICIENT: ranges from -1 to 1, with 0 indicating a symmetrical distribution, negative values indicating a left skew, and positive values indicating a right skew
df_SacDur['Saccade_Duration_bowleyCoef'] = (df_SacDur['Saccade_Duration_75%'] + df_SacDur['Saccade_Duration_25%'] - 2*df_SacDur['Saccade_Duration_med']) / df_SacDur['Saccade_Duration_DistQ1-Q3']

# drop columns not used in further analysis:
df_SacDur = df_SacDur.drop(['Saccade_Duration_25%', 'Saccade_Duration_75%'], axis=1)
df_SacDur
```
