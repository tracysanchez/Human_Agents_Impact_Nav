```python
import pandas as pd
# glob module for wildcards in paths
import glob
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import time
```

```python
start = time.time()
input_file = r"E:\WestbrookProject\SpaRe\control_group\pre-processing_2023\velocity_based\GazeProcessedMerge.csv"
chunk_size = 500000
total_rows = 21068033 #taken from df information (previous script)
csv_reader = pd.read_csv(input_file, chunksize=chunk_size) # low_memory = False
final_chunks = []
end = time.time()
print("Step 1 took:", end - start)

start = time.time()
for chunk in tqdm(csv_reader, total=total_rows // chunk_size + 1, desc="Reading"):
    first_column_name = chunk.columns[0]
    chunk[first_column_name] = chunk[first_column_name].astype(str)
    final_chunks.append(chunk)
df_full = pd.concat(final_chunks, ignore_index=True)

end = time.time()
print("Step 2 took:", end - start)
```
