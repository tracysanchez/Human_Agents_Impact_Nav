```python
import os
import pandas as pd
import glob
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "/Volumes/TwoTeras/0_Experiment_1/Proxemics/Minimal_Distance_Faces/" 

# csv files in the path
files = glob.glob(path + "/*.csv")
  
# defining an empty list to store 
# content
data_frame = pd.DataFrame()
content = []
  
# checking all the csv files in the 
# specified path
for filename in files:
    
    # reading content of csv file
    # content.append(filename)
    One_participant = pd.read_csv(filename)
    print(filename[-10:-4])
    content.append(One_participant)

# converting content to data frame
data_frame = pd.concat(content)
```

```python
# Create the 'Agent_Type' column based on the condition in ped_id
data_frame['Agent_Type'] = data_frame['ped_id'].apply(lambda x: 'Passive' if x.endswith('Sa') else 'Active' if x.endswith('Cma') else 'Unknown')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
Face = pd.read_csv("/Volumes/TwoTeras/0_Experiment_1/Proxemics/Minimal_Distances_Experiment_One_Faces.csv")
Body = pd.read_csv("/Volumes/TwoTeras/0_Experiment_1/Proxemics/Minimal_Distances_Experiment_One.csv")

# Set up the subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Create the distribution plot for Faces dataset on the left
sns.kdeplot(data=Face, x='minimal_distance', hue='Agent_Type', fill=True, palette="Set2", ax=axes[0])
axes[0].set_title('Distribution of Minimal Distance (Faces)', size=16)
axes[0].set_xlabel('Minimal Distance', size=14)
axes[0].set_ylabel('Density', size=14)

# Create the distribution plot for Body dataset on the right
sns.kdeplot(data=Body, x='minimal_distance', hue='Agent_Type', fill=True, palette="Set2", ax=axes[1])
axes[1].set_title('Distribution of Minimal Distance (Body)', size=16)
axes[1].set_xlabel('Minimal Distance', size=14)
axes[1].set_ylabel('Density', size=14)

# Adjust the layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Set up the subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Create the KDE plot for Faces dataset on the left
sns.kdeplot(data=Face, x='minimal_distance', hue='ped_id', fill=True, palette="Set2", ax=axes[0])
axes[0].set_title('Distribution of Minimal Distance (Faces)', size=16)
axes[0].set_xlabel('Minimal Distance', size=14)
axes[0].set_ylabel('Density', size=14)

# Create the KDE plot for Body dataset on the right
sns.kdeplot(data=Body, x='minimal_distance', hue='ped_id', fill=True, palette="Set2", ax=axes[1])
axes[1].set_title('Distribution of Minimal Distance (Body)', size=16)
axes[1].set_xlabel('Minimal Distance', size=14)
axes[1].set_ylabel('Density', size=14)

# Adjust the layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate the mean minimal distance for each ped_id for both datasets
order_face = Face.groupby('ped_id')['minimal_distance'].mean().sort_values().index
order_body = Body.groupby('ped_id')['minimal_distance'].mean().sort_values().index

# Set up the subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(25, 10))

# Create the box plot for the Faces dataset, ordered by mean minimal distance
sns.boxplot(data=Face, x='ped_id', y='minimal_distance', palette="Set2", order=order_face, ax=axes[0])
axes[0].set_title('Box Plot of Minimal Distance (Faces)', size=16)
axes[0].set_xlabel('Pedestrian ID', size=14)
axes[0].set_ylabel('Minimal Distance', size=14)
axes[0].tick_params(axis='x', rotation=90)

# Create the box plot for the Body dataset, ordered by mean minimal distance
sns.boxplot(data=Body, x='ped_id', y='minimal_distance', palette="Set2", order=order_body, ax=axes[1])
axes[1].set_title('Box Plot of Minimal Distance (Body)', size=16)
axes[1].set_xlabel('Pedestrian ID', size=14)
axes[1].set_ylabel('Minimal Distance', size=14)
axes[1].tick_params(axis='x', rotation=90)

# Adjust the layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate the mean minimal distance for each ped_id for both datasets
order_face = Face.groupby('SubjectID')['minimal_distance'].mean().sort_values().index
order_body = Body.groupby('SubjectID')['minimal_distance'].mean().sort_values().index

# Set up the subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(20, 10))

# Create the box plot for the Faces dataset, ordered by mean minimal distance
sns.boxplot(data=Face, x='SubjectID', y='minimal_distance', palette="Set2", order=order_face, ax=axes[0])
axes[0].set_title('Box Plot of Minimal Distance (Faces)', size=16)
axes[0].set_xlabel('Pedestrian ID', size=14)
axes[0].set_ylabel('Minimal Distance', size=14)
axes[0].tick_params(axis='x', rotation=45)

# Create the box plot for the Body dataset, ordered by mean minimal distance
sns.boxplot(data=Body, x='SubjectID', y='minimal_distance', palette="Set2", order=order_body, ax=axes[1])
axes[1].set_title('Box Plot of Minimal Distance (Body)', size=16)
axes[1].set_xlabel('Pedestrian ID', size=14)
axes[1].set_ylabel('Minimal Distance', size=14)
axes[1].tick_params(axis='x', rotation=45)

# Adjust the layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by SubjectID and Session to calculate the mean and std minimal distance for each person per session for both datasets
subject_session_stats_face = Face.groupby(['SubjectID', 'Session'])['minimal_distance'].agg(['mean', 'std']).reset_index()
subject_session_stats_face.rename(columns={'mean': 'mean_minimal_distance', 'std': 'std_minimal_distance'}, inplace=True)

subject_session_stats_body = Body.groupby(['SubjectID', 'Session'])['minimal_distance'].agg(['mean', 'std']).reset_index()
subject_session_stats_body.rename(columns={'mean': 'mean_minimal_distance', 'std': 'std_minimal_distance'}, inplace=True)

# Set up the subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot for the Face dataset
sns.lineplot(data=subject_session_stats_face, x='Session', y='mean_minimal_distance', hue='SubjectID', marker='o', ax=axes[0])
axes[0].set_title('Mean Minimal Distance per Person (Faces)', size=16)
axes[0].set_xlabel('Session', size=14)
axes[0].set_ylabel('Mean Minimal Distance', size=14)

# Plot for the Body dataset
sns.lineplot(data=subject_session_stats_body, x='Session', y='mean_minimal_distance', hue='SubjectID', marker='o', ax=axes[1])
axes[1].set_title('Mean Minimal Distance per Person (Body)', size=16)
axes[1].set_xlabel('Session', size=14)
axes[1].set_ylabel('Mean Minimal Distance', size=14)

# Adjust the layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
